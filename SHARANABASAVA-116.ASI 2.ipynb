```python
# 1️ Fix datatype (age to integer)
df['age'] = pd.to_numeric(df['age'], errors='coerce')
```

```python
# 2️ Handle missing numeric values using median
df['daily_screen_time'] = df['daily_screen_time'].fillna(
    df['daily_screen_time'].median()
)
```

```python
# 3️ Handle missing categorical values using mode
df['subscription_type'] = df['subscription_type'].fillna(
    df['subscription_type'].mode()[0])
```

```python
# 4️ Clean text columns (strip + title case)
text_cols = ['gender','subscription_type','device_type']
for col in text_cols:
    df[col] = df[col].str.strip().str.title()
```

```python
# 5️ Remove duplicate records
df = df.drop_duplicates()
```

```python
# 6️ Rename columns for clarity
df = df.rename(columns={
    'daily_screen_time': 'total_screen_time'
})
```

```python
# 7️ Simple outlier filtering (screen time > 16 hrs removed)
df = df[df['total_screen_time'] <= 16]
```

```python
# 1️ Total entertainment hours
df['total_entertainment_time'] = (
    df['netflix_hours'] +
    df['social_media_hours'] +
    df['music_hours'] +
    df['gaming_hours']
)
```

```python
# 3️ Age group category
df['age_group'] = pd.cut(
    df['age'],
    bins=[15,20,25,30,40],
    labels=['Teen','Young Adult','Adult','Senior']
)
```
